```python
#ANSWER 1:
a = 0

def b():
    global a
    a = c(a)

def c(a):
    return a + 2
```

Step by step execution:
1.Initial state:
    global variable a is initialized to 0.
2.First call: b():
    The function b() is called.
    Inside b(), global a is referring to the global variable a, not a local variable.
    The current value of a is 0(global).
    The function c(a) is called with a as the argument (0).
    Inside c(a), return a + 2 is evaluated, so it returns 0 + 2, which is 2.
    In b(), a is updated to 2 (the value returned by c(a)).
3.Second call: b():
    The function b() is called again.
    global a again refers to the global variable a, which is now 2.
    The function c(a) is called with a as the argument (2).
    Inside c(a), return a + 2 is evaluated, returning 2 + 2, which is 4.
    In b(), a is updated to 4.
4.Third call: b():
    The function b() is called.
    global a refers to the global variable a, which is now 4.
    The function c(a) is called with a as the argument (4).
    Inside c(a), return a + 2 is evaluated, returning 4 + 2, which is 6.
    In b(), a is updated to 6.
Final Value: a
After the three calls of b(), the value of a is now 6.

After first call: a becomes 2.
After second call: a becomes 4.
After third call: a becomes 6.
The final output a is 6.

```python
#ANSWER 3
class Marsupial:
    def _init_(self):
        self.pouch = []

    def put_in_pouch(self, item):
        """Add an item to the pouch."""
        self.pouch.append(item)

    def pouch_contents(self):
        """Return the contents of the pouch."""
        return self.pouch

class Kangaroo(Marsupial):
    def _init_(self, x, y):
        super()._init_()  # Initialize the parent Marsupial class
        self.x = x  # Kangaroo's x-coordinate
        self.y = y  # Kangaroo's y-coordinate

    def jump(self, dx, dy):
        """Move the kangaroo by dx along x-axis and dy along y-axis."""
        self.x += dx
        self.y += dy

    def _str_(self):
        """Return a string representation of the kangaroo's location."""
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"

# Example :
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())  # Output: ['doll', 'firetruck', 'kitten']

k = Kangaroo(0, 0)
print(k)  # Output: I am a Kangaroo located at coordinates (0,0)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())  # Output: ['doll', 'firetruck', 'kitten']
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)  # Output: I am a Kangaroo located at coordinates (3,0)
```

```python
#ANSWER 4 
def collatz(x):
    """Prints the Collatz sequence starting from x."""
    print(x)  # Print the current value of x
    if x == 1:
        return  # stop when x is 1
    elif x % 2 == 0:
        collatz(x // 2)  # Recursive call for even x
    else:
        collatz(3 * x + 1)  # Recursive call for odd x

# Example :
collatz(1)   # Output: 1
collatz(10)  # Output: 10, 5, 16, 8, 4, 2, 1
```

```python
#ANSWER 5
def binary(n):
    """Prints the binary representation of a non-negative integer n."""
    if n < 0:
        raise ValueError("Input must be a non-negative integer.")
    
    # Base case for 0
    if n == 0:
        print(0, end="")
    elif n == 1:
        print(1, end="")
    else:
        # Recursive call to binary for n // 2
        binary(n // 2)  # Print the binary representation of the quotient
        print(n % 2, end="")  # Print the remainder (0 or 1)

# Example :
binary(0)   # Output: 0
print()     # Print a newline for better output separation
binary(1)   # Output: 1
print()
binary(3)   # Output: 11
print()
binary(9)   # Output: 1001
print()
```

```python
#Created a new python file(heading_parser.py)

from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def _init_(self):
        super()._init_()
        self.headings = []  # List to store headings

    def handle_starttag(self, tag, attrs):
        """Handle start tags for headings (h1, h2, ..., h6)."""
        if tag.startswith('h') and tag[1].isdigit():  # Check if the tag is h1 to h6
            self.current_heading_level = int(tag[1])  # Get the level of heading

    def handle_data(self, data):
        """Handle the text data within the heading tags."""
        if hasattr(self, 'current_heading_level'):
            # Store the heading with its corresponding indentation level
            self.headings.append((self.current_heading_level, data.strip()))
            del self.current_heading_level  # Remove the current heading level after using it

    def print_headings(self):
        """Print the collected headings with the appropriate indentation."""
        for level, heading in self.headings:
            print(' ' * level + heading)  # Indent based on heading level

# Example :
if _name_ == "_main_":
    # Open and read the HTML file
    infile = open('w3c.html', 'r')
    content = infile.read()
    infile.close()

    # Create an instance of HeadingParser and feed it the HTML content
    hp = HeadingParser()
    hp.feed(content)

    # Print the extracted headings
    hp.print_headings()
```

```python
#Implementation of webdir function:

import requests
from bs4 import BeautifulSoup

def webdir(url, depth, indent):
    """Recursively prints URLs reachable from the given URL up to a certain depth."""
    if depth < 0:
        return  # Base case: if depth is negative, stop recursion
    
    # Print the current URL with the specified indentation
    print(' ' * indent + url)
    
    if depth == 0:
        return  # Base case: stop recursion if depth is zero
    
    try:
        # Make a request to the URL
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad responses
        
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all anchor tags and extract their href attributes
        for link in soup.find_all('a', href=True):
            # Create a full URL if the link is relative
            next_url = link['href']
            if not next_url.startswith('http'):
                next_url = requests.compat.urljoin(url, next_url)
            
            # Recursively call webdir for the next URL
            webdir(next_url, depth - 1, indent + 2)  # Increase indentation for the next level
            
    except requests.exceptions.RequestException as e:
        print(' ' * indent + f"Error: {e}")  # Print error message with indentation

```
